```python
import sys
sys.path.append('..')

import numpy as np
import pandas as pd

import warnings
warnings.filterwarnings('ignore')
```

```python
FILE_PATH = '../data/raw/zz800_close.CSV'
START_DATE = '2010-01-01'
END_DATE = '2020-12-31'
NAN_DROP_RATIO = 0.1
```

```python
def get_stock_data(
        file_path, 
        start_date, 
        end_date,
        nan_drop_ratio=0.1,
    ):
    start_date, end_date = pd.to_datetime(start_date), pd.to_datetime(end_date)
    df = pd.read_csv('../data/raw/zz800_close.CSV', encoding='gbk')
    df = df.rename(columns={'Unnamed: 0': 'date'})
    df = df.drop(columns=['Unnamed: 801'])
    df = df.drop([0, 1]).reset_index(drop=True)
    df['date'] = pd.to_datetime(df['date'].astype(int).astype(str))
    df = df[(df['date'] >= start_date) & (df['date'] <= end_date)].reset_index(drop=True)
    cols = (df.isnull().sum(0) / len(df) <= nan_drop_ratio).values
    df = df.iloc[:, cols]
    df = df.fillna(method='ffill').fillna(method='bfill')
    assert df.isnull().sum().sum() == 0
    df = df.reset_index(drop=True)
    return df
```

```python
df_stock = get_stock_data(FILE_PATH, START_DATE, END_DATE, NAN_DROP_RATIO)
df_stock
```
